```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
from pandas.plotting import autocorrelation_plot
```

```python
archivo = pd.read_csv("MonthDump.csv")
```

```python
archivo.head()
```

```python
splited = archivo.MES.str.split('/', expand=True)
```

```python
final = pd.concat([archivo, splited], axis=1, join='inner')
final.head()
```

```python
serie = pd.DataFrame({ 'A' : range(0, final.shape[0] ,1)})
```

```python
plt.figure(figsize=(50,20))
plt.plot(serie['A'],final['VENTAS'])
plt.show()
```

```python
plt.figure(figsize=(50,20))
plt.plot(serie['A'],final['CANTIDAD'])
plt.show()
```

```python
final['CANTIDAD'].autocorr()
```

```python
plt.figure(figsize=(50,50))
autocorrelation_plot(final['VENTAS'])
```
